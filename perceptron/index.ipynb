```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from Perceptron import Perceptron
import numpy as np
```

```python
X,y=make_blobs(n_samples=400,n_features=2,centers=2,cluster_std=1.3,random_state=1234)
Xt,Xtt,Yt,Ytt=train_test_split(X,y,test_size=0.2,random_state=1234)
```

```python
model=Perceptron()

model.fit(Xt,Yt)
```

```python
predictions=model.predict(Xtt)
```

```python
accuracy_score(Ytt,predictions)
```

```python
plt.scatter(Xt[:,0],Xt[:,1],marker='o',c=Yt)
plt.xlabel('Feat 1')
plt.ylabel('Feat 2')

min_feat1_x=np.min(Xt[:,0])
max_feat1_x=np.max(Xt[:,0])


# getting the second feat in terms of the first
y1=(model.weights[0]*min_feat1_x + model.bias)/(-model.weights[1])
y2=(model.weights[0]*max_feat1_x + model.bias)/(-model.weights[1])

plt.ylim([np.min(Xt[:,0])-3,np.max(Xt[:,1])+3])

plt.plot([min_feat1_x,max_feat1_x],[y1,y2])
plt.show()
```
